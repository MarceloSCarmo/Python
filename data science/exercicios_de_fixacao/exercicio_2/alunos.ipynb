```python
import pandas as pd

dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv')
```

```python
dados
```

```python
dados.isnull()

#   verificando se a base de dados tem algum dado nulo.
#   Como podemos ver, alguns outputs da coluna Notas retornam valores True, isso significa que existem dados nulos.
```

```python
dados.isnull().sum()

#   somando todos os dados nulos
```

```python
df_dados = dados.fillna(0)
df_dados

#   Tratando os dados de modo que os alunos com a coluna 'Notas' nula recebam a nota 0
#   Transformando valores nulos em 0
```

```python
#   Removendo os Alunos Alice e Carlos da base de dados por não fazerem mais parte da turma

dados.query('Nome == "Alice" | Nome == "Carlos"')
```

```python
dados.query('Nome == "Alice" | Nome == "Carlos"').index
```

```python
df_nomes_a_remover = dados.query('Nome == "Alice" | Nome == "Carlos"').index

#   Criando a variável para armazenar os nomes a serem removidos da base de dados
```

```python
df_dados.drop(df_nomes_a_remover, axis=0, inplace=True)

df_dados


#   Retornando o data frame com os nomes de Alice e Carlos excluidos da base de dados
```

```python
dados['Aprovado'] == 1
```

```python
aprovados = dados['Aprovado'] == 1
dados[aprovados]

#   Filtrando todos os alunos que foram aprovados
```

```python
nao_aprovados = df_dados['Aprovado'] == 0
df_dados[nao_aprovados]

#   Filtrando todos os alunos que não foram aprovados
```

```python
df_dados[aprovados].to_csv('alunos_aprovados.csv', index=False, sep=';')

#   Criando um arquivo csv somente com os alunos aprovados
```

```python
dados.query('Notas == 7')

#   Buscando os(as) alunos(as) com nota igual a 7
```

```python
df_dados.replace(to_replace= 7, value= 8 ) 

#   Utilizando o método replace para corrigir as notas erradas
```

Desafio Final

```python
df_dados['Pontos_extras'] = ( df_dados['Notas'] * 40 / 100 )

df_dados


#   Criando nova coluna "Pontos_extras"
```

```python
df_dados['Notas_finais'] = ( df_dados['Notas'] + df_dados['Pontos_extras'])

df_dados
```

```python
df_dados['Aprovado_final'] = df_dados['Notas_finais'].apply(lambda x: 'True' if x >= 6 else 'False') 
```
